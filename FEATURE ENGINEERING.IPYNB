```python
df['Urban_Males'] = df['Main Workers - Urban - Males']
df['Urban_Females'] = df['Main Workers - Urban - Females']
df['Rural_Males'] = df['Main Workers - Rural - Males']
df['Rural_Females'] = df['Main Workers - Rural - Females']
```

```python
industry_mapping = {
    'Retail': 'Consumer Goods',
    'Poultry': 'Agriculture',
    'Manufacturing': 'Manufacturing'
}
```

```python
df['Industry_Group'] = df['NIC Name'].map(industry_mapping).fillna('Other')
```

```python
df['Total_Workers'] = df['Main Workers - Total -  Persons'] + df['Marginal Workers - Total -  Persons']
```

```python
df['MaleFemale_Ratio'] = df['Main Workers - Total - Males'] / (df['Main Workers - Total - Females'].replace(0, 1))
```

```python
df[['Total_Workers','MaleFemale_Ratio']].isnull().sum()
```

```python
df[['Total_Workers','MaleFemale_Ratio']] = df[['Total_Workers','MaleFemale_Ratio']].astype(int)
```

```python
df[['State', 'District']] = df['India/States'].str.extract(r'STATE - (\w+)|District - (\w+)', expand=True)
df = df.drop('India/States', axis=1)
```

```python
df[['State', 'District']].isnull().sum()
```

```python
df['State'] = df['State'].ffill()
df['District'] = df['District'].bfill()
```

```python
df[['State', 'District']].isnull().sum()
```
